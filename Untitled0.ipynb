# City functions

```python
import numpy as np
from collections import deque
import random
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import ListedColormap

# Define a class for the buildings
class Building:
    def __init__(self, name, width, height, requires_road=False):
        self.name = name
        self.width = width
        self.height = height
        self.requires_road = requires_road

# Function to check if a building can fit in a given position
def can_place_building(city, building, x, y):
    for i in range(building.width):
        for j in range(building.height):
            if x + i >= city_width or y + j >= city_height:
                return False
            if (city[y + j][x + i] is not None) and (city[y + j][x + i] != "R"):
                return False
    return True

# Function to place a building on the city grid
def place_building(city, building, x, y):
    for i in range(building.width):
        for j in range(building.height):
            city[y + j][x + i] = building.name[0]

# Function to remove a building from the city grid
def remove_building(city, building, x, y):
    for i in range(building.width):
        for j in range(building.height):
            city[y + j][x + i] = None

# Function to visualize the city or connectivity map
def print_city(city):
    for row in city:
        if isinstance(row[0], int):
            print(" ".join([str(cell) for cell in row]))
        else:
            print(" ".join(['.' if cell is None else cell[0] for cell in row]))

def plot_city(city, figsize=(12,10)):
    """
    Plots the city grid as a colored matrix with a legend.

    :param city: 2D numpy array representing the city grid.
    """
    # Define the mapping from cell types to integer codes, colors, and labels
    cell_type_mapping = {
        'C': {'code': 3, 'color': '#ff7f0e', 'label': 'City Hall'},        # Orange
        'G': {'code': 0, 'color': '#d62728', 'label': 'Galeria'},          # Red
        'c': {'code': 1, 'color': '#2ca02c', 'label': 'Campo de Soja'},    # Green
        'H': {'code': 2, 'color': '#1f77b4', 'label': 'House'},            # Blue
        'S': {'code': 4, 'color': '#9467bd', 'label': 'Santuário'},        # Purple
        'R': {'code': 5, 'color': '#8c564b', 'label': 'Road'},             # Brown
        'B': {'code': 6, 'color': '#7f7f7f', 'label': 'Blocked'},          # Gray
        '.': {'code': 7, 'color': '#ffffff', 'label': 'Empty'},            # White
        None: {'code': 7, 'color': '#ffffff', 'label': 'Empty'}            # White
    }

    # Initialize the numeric grid
    numeric_grid = np.zeros(city.shape, dtype=int)

    # Iterate and map cell types to codes
    for y in range(city.shape[0]):
        for x in range(city.shape[1]):
            cell = city[y][x]
            if cell in cell_type_mapping:
                code = cell_type_mapping[cell]['code']
            else:
                code = cell_type_mapping[None]['code']
            numeric_grid[y][x] = code

    # Define the colormap based on the cell type mapping
    cmap = ListedColormap([
        cell_type_mapping['G']['color'],   # 0: Galeria
        cell_type_mapping['c']['color'],   # 1: Campo de Soja
        cell_type_mapping['H']['color'],   # 2: House
        cell_type_mapping['C']['color'],   # 3: City Hall
        cell_type_mapping['S']['color'],   # 4: Santuário
        cell_type_mapping['R']['color'],   # 5: Road
        cell_type_mapping['B']['color'],   # 6: Blocked
        cell_type_mapping['.']['color']    # 7: Empty
    ])

    # Create the plot
    plt.figure(figsize=figsize)
    plt.imshow(numeric_grid, cmap=cmap, aspect='equal')

    # Create custom legend
    legend_patches = []
    added_labels = set()
    for cell, props in cell_type_mapping.items():
        label = props['label']
        color = props['color']
        if label not in added_labels:
            patch = mpatches.Patch(color=color, label=label)
            legend_patches.append(patch)
            added_labels.add(label)

    plt.legend(handles=legend_patches, bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

    # Optionally, add text labels to each cell for clarity
    for y in range(city.shape[0]):
        for x in range(city.shape[1]):
            cell = city[y][x]
            if cell is not None and cell != '.':
                plt.text(x, y, cell, ha='center', va='center', color='black', fontsize=8)

    # Remove axis ticks for a cleaner look
    plt.xticks([])
    plt.yticks([])

    plt.title('City Layout Visualization')
    plt.tight_layout()
    plt.show()

# Function to check if a position is connected to City Hall via a road
def get_connected_map(city):
    connected_map = [[0 for _ in range(city_width)] for _ in range(city_height)]
    queue = []
    # Find initial connected roads adjacent to City Hall
    for y in range(city_height):
        for x in range(city_width):
            if city[y][x] == 'C':
                for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                    nx, ny = x + dx, y + dy
                    if 0 <= nx < city_width and 0 <= ny < city_height and city[ny][nx] == 'R':
                        queue.append((nx, ny))
    # BFS to find all connected roads
    while queue:
        cx, cy = queue.pop(0)
        connected_map[cy][cx] = 1
        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            nx, ny = cx + dx, cy + dy
            if 0 <= nx < city_width and 0 <= ny < city_height and connected_map[ny][nx] == 0:
                if city[ny][nx] == 'R':
                    queue.append((nx, ny))
                else:
                  connected_map[ny][nx] = 1
    return connected_map

def is_building_connected(building, x, y, connected_map):
    for i in range(building.width):
        for j in range(building.height):
            if connected_map[y + j][x + i] == 1:
                return True
    return False

# Define a class for an individual solution
class Individual:
    def __init__(self, city, building_coords=None):
        self.city = city.copy()
        self.building_coords = building_coords if building_coords is not None else []

    def initialize_roads(self):
        for y in range(city_height):
            for x in range(city_width):
                if self.city[y][x] is None:
                    self.city[y][x] = 'R' if random.random() < 0.5 else None

    def initialize_buildings(self, buildings):
        self.building_coords = []
        for building in buildings:
            x = random.randint(0, city_width - building.width)
            y = random.randint(0, city_height - building.height)
            self.building_coords.append((building, x, y))

def has_unconnected_buildings(city, individual):
    unconnected_buildings = []
    connected_map = get_connected_map(city)
    for (building, x, y) in individual.building_coords:
        if not is_building_connected(building, x, y, connected_map):
            unconnected_buildings.append(building)
    return(unconnected_buildings)
```

# Genetic algorithm functions

```python
import matplotlib.pyplot as plt
from scipy.ndimage import label

def fit(individual, penalize_many_roads=True, penalize_sparce_free_spaces=True):
    fitness = 0
    city = individual.city.copy()

    # Penalize based on unplaced buildings
    for (building, x, y) in individual.building_coords:
        if can_place_building(city, building, x, y):
            place_building(city, building, x, y)
        else:
            fitness -= 10

    # Penalize based on unconnected buildings
    unconnected_buildings = has_unconnected_buildings(city, individual)
    for building in unconnected_buildings:
        fitness -= 1

    # Penalize based on the number of roads in the city
    if penalize_many_roads:
        num_roads = sum(cell == 'R' for row in city for cell in row)
        fitness -= 0.01 * num_roads

    # Penalize based on scattered free spaces
    if penalize_sparce_free_spaces:
        free_space = (city == None)
        labeled_array, num_features = label(free_space)
        if num_features > 1:
            fitness -= (num_features - 1) * 0.01

    return fitness

# Function to generate an initial population of cities
def generate_initial_population(pop_size, city, buildings):
    population = [Individual(city) for _ in range(pop_size)]
    for individual in population:
        individual.initialize_buildings(buildings)
        individual.initialize_roads()
    return population

def get_parent_probabilities(population, fitnesses, exponent = 1.2):
    probabilities = exponent**np.array(fitnesses) / np.sum(exponent**np.array(fitnesses))
    return probabilities

# Select individuals based on their fitness
def select_parents(population, probabilities, num_parents=2):
    selected_indices = np.random.choice(len(population), size=num_parents, p=probabilities, replace=False)
    return [population[i] for i in selected_indices]

# Mutation function to introduce variation in an individual
def mutate(individual, building_mutation_rate=0.05, mean_road_mutation=2):
    # Mutate building positions
    for i in range(len(individual.building_coords)):
        if random.random() < building_mutation_rate:
            building, x, y = individual.building_coords[i]
            new_x = max(0, min(city_width - building.width, x + random.choice([-1, 0, 1])))
            new_y = max(0, min(city_height - building.height, y + random.choice([-1, 0, 1])))
            individual.building_coords[i] = (building, new_x, new_y)

    # Mutate roads by choosing random cells to mutate
    num_road_mutations = np.random.poisson(mean_road_mutation)
    for _ in range(num_road_mutations):
        x = random.randint(0, city_width - 1)
        y = random.randint(0, city_height - 1)
        if individual.city[y][x] == 'R' or individual.city[y][x] is None:
            individual.city[y][x] = 'R' if random.random() < 0.5 else None

# Crossover function to combine two parents into an offspring
def crossover(parent1, parent2):
    # Crossover for building coordinates
    child_buildings = []
    for i in range(len(parent1.building_coords)):
        if random.random() < 0.5:
            child_buildings.append(parent1.building_coords[i])
        else:
            child_buildings.append(parent2.building_coords[i])

    # Combine road layouts from both parents
    child_city = np.where(np.random.rand(*parent1.city.shape) > 0.5, parent1.city, parent2.city)

    return Individual(child_city, child_buildings)

def generate_random_individual(city, buildings):
    individual = Individual(city)
    individual.initialize_buildings(buildings)
    individual.initialize_roads()
    return individual

# Main evolve function
def main_evolve(city, buildings, num_population              = 10,
                                 num_generations             = 100,
                                 building_mutation_rate      = 0.05,
                                 mean_road_mutation          = 2,
                                 parent_probability_exponent = 1.2,
                                 elite_percentage            = 0.05,   # 5% elitism
                                 new_individual_percentage   = 0.20,   # 20% new individuals
                                 penalize_many_roads         = True,
                                 penalize_sparce_free_spaces = True,
                                 individual                  = None):
    # Initial function call to generate a population
    population = generate_initial_population(num_population, city, buildings)
    best_fitnesses = []
    mean_fitnesses = []  # List to store mean fitness per generation
    best_individual = None

    if individual is not None:
        population.append(individual)

    for generation in range(num_generations):
        # Evaluate fitness for all individuals in the population
        fitnesses = [fit(individual, penalize_many_roads, penalize_sparce_free_spaces) for individual in population]
        max_fitness = max(fitnesses)
        mean_fitness = sum(fitnesses) / len(fitnesses)  # Calculate mean fitness
        best_fitnesses.append(max_fitness)
        mean_fitnesses.append(mean_fitness)  # Store mean fitness

        # Track the best individual
        if best_individual is None or max_fitness > fit(best_individual, penalize_many_roads, penalize_sparce_free_spaces):
            print("New best! fitness:", max_fitness)
            best_individual = population[np.argmax(fitnesses)]

        # Uncomment the following line to print fitness details each generation
        # print(f"Generation {generation}: Best Fitness = {max_fitness:.2f}, Mean Fitness = {mean_fitness:.2f}")

        new_population = []

        # **Elitism: Select the top elite_percentage of individuals**
        num_elites = max(1, int(np.ceil(num_population * elite_percentage)))  # Ensure at least one elite
        population_fitness_pairs = list(zip(population, fitnesses))
        sorted_population = sorted(population_fitness_pairs, key=lambda x: x[1], reverse=True)
        elites = [individual for individual, fitness in sorted_population[:num_elites]]
        new_population.extend(elites)

        # **Add new random individuals to the population**
        for _ in range(int(np.ceil(num_population * new_individual_percentage))):
            new_individual = generate_random_individual(city, buildings)
            new_population.append(new_individual)

        # **Generate new individuals via selection, crossover, and mutation**
        parent_probabilities = get_parent_probabilities(population, fitnesses, parent_probability_exponent)
        while len(new_population) < num_population:
            parents = select_parents(population, parent_probabilities, num_parents=2)
            kid = crossover(parents[0], parents[1])
            mutate(kid, building_mutation_rate, mean_road_mutation)
            new_population.append(kid)

        population = new_population

    return best_individual, best_fitnesses, mean_fitnesses
```
